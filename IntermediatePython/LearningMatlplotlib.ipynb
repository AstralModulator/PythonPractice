```python
import matplotlib.pyplot as plt
```

```python
year = [1950,1970,1990,2010]
```

```python
population = [2.51944,3.692,5.263,6.972]
```

```python
plt.figure(figsize=(16,6))
plt.plot(year, population)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Year vs Population')
plt.show()
```
