```python
df['life_style_score'] = df.freetime + df.goout + df.Dalc + df.Walc
```

```python
df['life_style_score'] = (df['life_style_score'] - df['life_style_score'].min())/(df['life_style_score'].max()-df['life_style_score'].min())
```

```python
df['life_style_score']
```

```python
df['life_style_score'].corr(df.G3)
```

```python
upper10 = df.nlargest(int(0.25 * len(df)),'life_style_score')['G3'].mean()
```

```python
bottom10 = df.nsmallest(int(0.25 * len(df)),'life_style_score')['G3'].mean()
```
