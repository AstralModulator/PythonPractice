```python
df = pd.read_csv(r"C:\Users\HP\Downloads\student_data.csv")
```

```python
df['parent_edu_avg'] = (df['Medu'] + df['Fedu']) / 2
```

```python
df['parent_edu_level'] = pd.cut(
    df['parent_edu_avg'],
    bins=[0, 1.5, 3, 4],
    labels=['Low', 'Medium', 'High']
)
```

```python
df.groupby('parent_edu_level')['G3'].mean().sort_values(ascending=False)
```

```python
df['lifestyle_score'] = (
    df['freetime'] + df['goout'] + df['Dalc'] + df['Walc']
)
```

```python
df['lifestyle_score_norm'] = (
    (df['lifestyle_score'] - df['lifestyle_score'].min()) /
    (df['lifestyle_score'].max() - df['lifestyle_score'].min())
)
```

```python
df['lifestyle_score_norm'].corr(df['G3'])
```

```python
top_fun = df.nlargest(int(0.25 * len(df)), 'lifestyle_score_norm')['G3'].mean()
low_fun = df.nsmallest(int(0.25 * len(df)), 'lifestyle_score_norm')['G3'].mean()
top_fun, low_fun
```

```python
df.assign(passed=df['G3'] >= 10) \
  .groupby('failures')['passed'] \
  .mean()
```
